## Q4 - Can you detect cascading failures as delays in one airport create delays in others?

This is a challenging question. First I will find the most occuring TailNum in a single day within my data to track if a flight displays cascading delays as it flies from airport to airport through out the day.

```python
delays.loc[:, ['Year','Month','DayofMonth','TailNum']].value_counts()
```

This is sad, the most occuring trips in a day by the same flight within the subset is only 2 by Flight with TailNum - N13970. 

```python
delays.loc[(delays['TailNum'] == 'N13970'), ['TailNum','Year','Month','DayofMonth','DayOfWeek','DepTime','DepDelay','Origin','ArrTime',
                                            'Dest','ArrDelay']].sort_values('Year')
```

Taking a closer look at the data, this flight was not delayed on 1st August, 2003. Infact we can see that it leaves early from ICT at 6:50 am in the morning and then reaches early by 8 mins at IAH airport. Again  it leaves 5 mins early from HRL airport and reaches IAH airport 13 mins early. Between the two trips in the data the flight N13970 would have made a trip from IAH to HRL.We can conclude that the middle trip from IAH to HRL was not delayed due to the early Departure of trip 3. 
Hence, from this short data it looks like there is a cascade of non-delays between airports that is if a flight is not at one airport it is not delayed at the other. However, this is too little evidence. 

I will now take a look at the day 1st August 2003 on which day N13970 has 2 records. 

```python
delays.loc[(delays['Year'] == 2003) & (delays['Month'] == 8) &(delays['DayofMonth'] == 1), ['TailNum',
                                        'DepTime','DepDelay','Origin','ArrTime','Dest','ArrDelay',
                                        'ActualElapsedTime','CRSElapsedTime']].sort_values('DepTime')
```

On examining data on 1st August, 2003, we can see that most flights were not delayed on that day. There is also a pattern where if the flight departs early it arrives early. Cascading early arrivals.

I will now look for a better solution to this problem by examining the day that has the most number of enteries in the data. 

```python
delays.loc[:, ['Year','Month','DayofMonth']].value_counts().head(10)
```

4th October 2004 has 14 rows in my data. Let me see if I can find any evidence for cascading events.

```python
delays.loc[(delays['Year'] == 2004) & (delays['Month'] == 10) &(delays['DayofMonth'] == 4), ['Year','Month','DayofMonth',
                                        'DepTime','DepDelay',
                                        'Origin','ArrTime','TailNum','Dest','ArrDelay',
                                        ]].sort_values('DepTime').reset_index()
```

By looking at the first 2 rows trip 0(index no) from ABQ to DEN at 6:36 am is not delayed leading to an early arrival in DEN at 7:47 am. 

Trip 1 from DEN at 8:00 am is not delayed, it also arrives at CLT airport 6 mins early at 12:50pm. 

Trip 6 leaving from CLT later in the day at 14:21pm is early by 4 mins.

Flights in the later part of the day get delayed and we can notice a pattern where flights that take off late, also land late. 

By this method too we can see a pattern of early departure lead to early arrivals and delayed departure lead to delayed arrivals. 

One last try to see if I can detect cascading failures. 

```python
q4 = delays.loc[delays['Delayed'] == 1, ['Year','Month','DayofMonth','TailNum','DepTime','Origin','DepDelay','ArrTime','Dest',
                                   'ArrDelay']].sort_values(by = ['Year','Month','DayofMonth','DepTime'])
q4.head()
```

In the above cell I have selected my data and arranged them in the time sequence that they occur. Now I will create a Lagged column that holds the Lagged variable of the Dest column. By doing this I will be able to get the previous destination airport and the next Origin airport on the same row. 

```python
#creating a lagged variable to make a sequence in time
q4["Lagged"] = q4["Dest"].shift(1)
q4.head()
```

Selecting columns where Lagged == Origin. This way I can see if I have flights that land and take off from the same airport on after the other. 

```python
#Selecting data where the Origin airport is the same as the preivous Dest airport
q4_a = q4.loc[q4['Origin'] == q4['Lagged'], ['Year','Month','DayofMonth','TailNum',
                                      'DepTime','Origin','DepDelay','ArrTime','Dest',
                                   'ArrDelay','Lagged']].sort_values(by = ['Year','Month','DayofMonth','DepTime']).reset_index()

#couting which day has the most number of rows where Origin airport is same as the previous Dest Airport 
q4_a.loc[:,['Year','Month','DayofMonth']].value_counts().head()
```

```python
q4_a.loc[(q4_a['Year'] == 2004) & (q4_a['Month'] == 12) & (q4_a['DayofMonth'] == 6), ['Year','Month','DayofMonth','TailNum',
                                        'DepTime','DepDelay','Origin','ArrTime','Dest','ArrDelay',
                                        ]].sort_values('DepTime')
```

From the above data on 6th Dec, 2004 we can see that there is an ArrDelay at 7:40am at SMF airport and a corresponding DepDelay at 11:05am from SMF airport. This is a case of cascading failures from one airport to the other.

Whereas the data below for 14th July, 2005 is not in the correct time sequence so nothing can be concluded.

```python
q4_a.loc[(q4_a['Year'] == 2005) & (q4_a['Month'] == 7) & (q4_a['DayofMonth'] == 14), ['TailNum',
                                        'DepTime','DepDelay','Origin','ArrTime','Dest','ArrDelay',
                                        ]].sort_values('DepTime')
```

To conclude the data provides glimpses of cascading delays from one airport to another but this is not sufficient evidence to prove cascading delays.

```python
numerical_features = ['DepDelay','Distance','TaxiIn','TaxiOut','CRSElapsedTime','CarrierDelay',
                     'NASDelay','WeatherDelay','SecurityDelay','LateAircraftDelay']

# Applying SimpleImputer and StandardScaler into a pipeline
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer()),
    ('scaler', StandardScaler())])

categorical_features = ['Cancelled', 'Diverted', 'Year','Month','DayofMonth','DayOfWeek','DepHour']

# Applying SimpleImputer and then OneHotEncoder into another pipeline
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer()),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

data_transformer = ColumnTransformer(
    transformers=[
        ('numerical', numerical_transformer, numerical_features),
        ('categorical', categorical_transformer, categorical_features)]) 
```
